**(모자이크 플롯 & 오즈비)**

```r
library(mosaic)
```

```r
table <- tally(~ god_imp + happy, data = life)   ## 2차원 분할표
mosaicplot(table, color=c("Navy", "Brown"), main="신의 중요성 여부에 따른 행복감")
```

- 삶에서 신이 중요하다고 답한 사람이 행복감을 느끼는 비율(brown)이 신이 중요하지 않다고 답한 사람이 행복감을 느끼는 비율(brown)보다 컸다. 

```r
odds1 = table[2,2]/table[2,1] ; odds1 ## 신을 중요하게 여기는 집단이 행복감을 느낄 오즈
odds2 = table[1,2]/table[1,1] ; odds2 ## 신을 중요하지 않게 여기는 집단이 행복감을 느낄 오즈
odds1/odds2  ##OR
```

- 신을 중요하게 여기는 집단에서 행복감을 느낄 오즈가 신을 중요하게 여기지 않는 집단에서 행복감을 느낄 오즈보다 약 1.87배 높다. 

```r
table <- tally(~ god_deep + happy, data = life)   ## 2차원 분할표
mosaicplot(table, color=c("Navy", "Brown"), main="신의 중요성 여부에 따른 행복감")
```

- 유신론자가 행복감을 느끼는 비율(brown)과 무신론자가 행복감을 느끼는 비율(brown)은 비슷해보인다.

```r
odds1 = table[2,2]/table[2,1] ; odds1 ## 유신론자기 행복감을 느낄 오즈
odds2 = table[1,2]/table[1,1] ; odds2 ## 무신론자 행복감을 느낄 오즈
odds1/odds2  ##OR
```

- 유신론자가 행복감을 느낄 오즈가 무신론자가 행복감을 느낄 오즈보다 약 1.34배 높다. 

```r
table <- tally(~ trust + happy, data = life)   ## 2차원 분할표
mosaicplot(table, color=c("Navy", "Brown"), main="타인 신뢰 여부에 따른 행복감")
```

- 타인을 신뢰하는 사람이 행복감을 느끼는 비율(brown)이 타인을 신뢰하지 않는 사람이 행복감을 느끼는 비율(brown)보다 컸다. 

```r
odds1 = table[2,2]/table[2,1] ; odds1 ## 타인을 신뢰하는 집단에서 행복감을 느낄 오즈
odds2 = table[1,2]/table[1,1] ; odds2 ## 타인을 신뢰하지 않는 집단에서 행복감을 느낄 오즈
odds1/odds2  ##OR
```

- 타인을 신뢰하는 집단에서 행복감을 느낄 오즈가 타인을 신뢰하지 않는 집단에서 행복감을 느낄 오즈보다 약 2배 높다. 

```r
table <- tally(~ inside + happy, data = life)   ## 2차원 분할표
mosaicplot(table, color=c("Navy", "Brown"), main="외향성에 따른 행복감")
```

- 외향적인 사람이 행복감을 느끼는 비율이 외향적이지 않은 사람이 행복감을 느끼는 비율보다 컸다. 

```r
odds1 = table[2,2]/table[2,1] ; odds1 ## 외향적인 집단에서 행복감을 느낄 오즈
odds2 = table[1,2]/table[1,1] ; odds2 ## 외향적이지 않은 집단에서 행복감을 느낄 오즈
odds1/odds2  ##OR
```

- 외향적인 집단에서 행복감을 느낄 오즈가 외향적이지 않은 집단에서 행복감을 느낄 오즈보다 약 2.26배 높다. 

```r
table <- tally(~ comp_mean + happy, data = life)   ## 2차원 분할표
mosaicplot(table, color=c("Navy", "Brown"), main="경쟁의 인식에 따른 행복감")
```

```r
odds1 = table[2,2]/table[2,1] ; odds1 ## 경쟁이 유익하다고 인식하는 사람들이 행복감을 느낄 오즈
odds2 = table[1,2]/table[1,1] ; odds2 ## 경쟁이 유익하지 않다고 인식하는 사람들이 행복감을 느낄 오즈
odds1/odds2  ##OR
```

- 경쟁이 유익하다고 인식하는 사람들이 행복감을 느낄 오즈가 경쟁이 유익하지 않다고 인식하는 사람들이 행복감을 느낄 오즈보다 약 0.63배이다.

### **모형 적합**
- 연구 문제와 탐색적 자료 분석의 결과를 참고하여 모형 적합

**Model1**  `god_imp` 설명변수

```r
model1 <- glm(happy ~ god_imp , family = binomial, data = life)
summary(model1)
```

```r
exp(model1$coefficients)
exp(confint(model1))
```

$$log(\frac{\hat{p}}{1-\hat{p}}) = \beta_0 + \beta_1*god.imp = 1.48710 +  0.62683*god.imp$$

$$\frac{\hat{p}}{1-\hat{p}} = e^{1.48710 +  0.62683*god.imp}$$

- $\hat{p}$ = 삶에 행복감을 느끼는 사람의 추정 비율
- 삶에 행복감을 느끼는 신을 중요히 여기는 사람의 오즈는 신을 중요히 여기지 않는 사람의 $e^{0.62683} = 1.87$ 배($p-value =0.00023$ )
- 오즈비의 95% 신뢰구간 = (1.35, 2.63) 으로 1을 포함하지 않으므로 위 설명 변수는 유의하다. 

**Model2**  `god_imp`, `pray_n` ,`god_imp*pray_n`가 설명변수

```r
model2 <- glm(happy ~ pray_n + god_imp + pray_n:god_imp , family = binomial, data = life)
summary(model2)
```

- 기도횟수와 신을 중요히 여기는지 여부에 따른 기도횟수의 교호작용을 포함한 회귀모형을 적합함
- 기도횟수와 교호작용 항은 유의하지 않음
- 기도횟수에 따라 행복감을 느끼는 여부는 다르지 않음
- 신을 중요히 여기는지 여부에 따라 기도횟수는 다르지 않음 <br>
-> 최종 모형에 `pray_n`과 `pray_n*god_imp`는 미포함

**Model3**  `god_imp`, `pray_n` , `god_imp*pray_n`, `trust` , `inside` ,`comp_mean`가 설명변수

```r
model3 <- glm(happy ~ pray_n + god_imp + pray_n:god_imp +  trust + inside + comp_mean  , family = binomial, data = life)
summary(model3)
```

```r
exp(model3$coefficients)
exp(confint(model3))
```

- `trust`, `inside`,`god_imp` , `comp_mean`는 신뢰구간에 1을 포함하지 않으므로 유의한 설명변수이다. ($p<0.1$)

**Model4**  `god_imp`, `pray_n` , `god_imp*pray_n`, `trust` , `inside` , `poli` , `poli_parti`, `god_deep`, `comp_mean`가 설명변수

```r
model4 <- glm(happy ~ pray_n + god_imp + pray_n:god_imp +  trust + inside +poli + poli_parti + god_deep+comp_mean, family = binomial, data = life)
summary(model4)
```

- `poli`,`poli_parti`, `god_deep` 변수가 유의하지 않기 때문에 삶의 행복감을 느끼는 여부는 정치적 요인과 무관하다고 생각한다. 

**Drop-in-deviance Tests/AIC**

```r
anova(model1, model2,model3,model4 , test = "Chisq")
AIC(model1, model2,model3,model4)
```

- model2와 model3을 비교할 때, drop-in-deviance는 1045.585 -1006.265 = 39.32
    - 자유도 2의 $X^2$-검정에서 $p$-value= 0으로 두 모형의 차이가 매우 유의하다. 
- model3에서 model4의 설명변수 추가는 도움이 되지 않는다. ($p$-value = 0.79) <br>
-> 가장 유의확률이 작은 model3의 성능이 좋다. 
- AIC도 함께 비교해봤을 때 AIC값이 가장 작은 model3가 가장 좋다. 

**(최종 모형)** Model5 :  Model3에서 유의하지 않은 변수 제거

```r
model5 <- glm(happy ~  god_imp +  trust + inside +comp_mean , family = binomial, data = life)
summary(model5)
```

$$log(\frac{\hat{p}}{1-\hat{p}})=  1.2017 +  0.5123*god.imp +   0.6710*trust + 0.7136*inside -0.3742*comp.mean $$ <br>
$$\frac{\hat{p}}{1-\hat{p}}=  e^{1.2017 +  0.5123*god.imp +   0.6710*trust + 0.7136*inside -0.3742*comp.mean } $$

```r
exp(model5$coefficients)
exp(confint(model5))
```

- model3에서 유의하지 않은 변수를 제거하여 model5를 적합하였다.
- `trust`, `inside`,`god_imp`는 신뢰구간에 1을 포함하지 않으므로 유의한 설명변수이다. ($p<0.01$)
- 다른 설명변수를 통제한 후 신을 중요히 여기는 사람이 삶에 행복감을 느낄 오즈는 신을 중요히 여기지 않는 사람이 삶에 행복감을 느낄 오즈의 $e^{ 0.5123} = 1.67$ 배 
- 다른 설명변수를 통제한 후 외향적인 사람이 삶에 행복감을 느낄 오즈는 외향적이지 않은 사람이 삶에 행복감을 느낄 오즈의 $e^{0.7136} =2.04$ 배
- 다른 설명변수를 통제한 후 타인을 신뢰하는 사람이 삶에 행복감을 느낄 오즈는 타인을 신뢰하지 않은 사람이 삶에 행복감을 느낄 오즈의 $e^{ 0.6710} =1.96$ 배
- 다른 설명변수를 통제한 후 경쟁이 유익하다고 생각하는 사람이 삶에 행복감을 느낄 오즈는 경쟁이 유익하다고 생각하지 않는 사람이 삶에 행복감을 느낄 오즈의 $e^{-0.3742} =0.69$ 배

**Drop-in-deviance Tests/AIC**

```r
anova(model5, model3 , test = "Chisq")
AIC(model5, model3)
```

- model3와 model5을 비교할 때, drop-in-deviance는  0.4186924
    - 자유도 2의 $X^2$-검정에서 $p$-value= 0.8으로 두 모형의 차이가 거의 없다. 
-> model5에서 model3의 설명변수 추가는 도움이 되지 않는다.<br>
-> 설명변수의 개수가 더 적은 model5의 성능이 좋다. 
- AIC도 함께 비교해봤을 때 AIC값이 가장 작은 model5가 가장 좋다.
